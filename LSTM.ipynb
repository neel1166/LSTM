```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
sales_data=pd.read_csv('Sales_Transactions_Dataset_Weekly.csv')
```

```python
sales_data_norm  = sales_data.iloc[:,55:106] # LSTM training data
week51 = sales_data['Normalized 51'] #Target feature
rows,column = sales_data_norm.shape
```

```python
sales_data_norm.head()
```

```python
window = 7 # LSTM window size
pred_week51 = np.array([]) # preicted data of week 51

# Building LSTM encoder and prediction of data using Ridge Linear Clssifier
for j in range(rows):
    train=[] # LSTM window size training labels
    labels =[]
    r =sales_data_norm.iloc[j].values 
    lstm_mat=np.asarray([r[i:i+window] for i in range(0, len(r), 1)])
    lstm_mat=lstm_mat[:-(window-1)]
    for i in range(len(lstm_mat)-1):
        train.append(lstm_mat[i][0:window-1])
        labels.append(lstm_mat[i][-1])
    test=[lstm_mat[-1][1:window]]
    #classification
    clf = Ridge(alpha=1.0)
    clf.fit(train,labels)
    prediction = clf.predict(test).flat[0]
    prediction = np.round(prediction,2)
    pred_week51 = np.append(pred_week51,prediction)

pred_week51 = np.where(pred_week51 < 0, 0, pred_week51)

mean_squared_error(week51.values,pred_week51) * 100
```
